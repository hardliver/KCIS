```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame(
    [0.58, 0.18, 0.16, 0.07, 0.01],
    index=['a=58%', 'b=18%', 'c=16%', 'd=7%', 'e=1%'],
    columns=['']
)
```

```python
df.plot(
    kind='pie',
    subplots=True,
    figsize=(4,4)
)
```

```python
def render_index(data):
    index = []
    for d,a in zip(data, string.ascii_lowercase):
        tmp_index = f'{a}={d}%'
        index.append(tmp_index)
    return index
```

```python
objects = ('A', 'B', 'C', 'D', 'E', 'F')
x_pos = np.arange(len(objects))
Rfrequency = [0.51, 0.19, 0.09, 0.12, 0.06, 0.03]
```

```python
plt.bar(x_pos, Rfrequency, align='center', alpha=0.5, color='red')
plt.xticks(x_pos, Rfrequency)
plt.ylabel('Relative frequency')
plt.title('Bar chart')
```
