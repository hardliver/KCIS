```python
# 導入 requests 和 lxml 模組
import requests
import lxml.html
```

```python
url = 'https://mingyanjiaju.org/xiehouyu/40033.html'
resp = requests.get(url)  # 透過 GET 方法獲取數據
```

```python
# 檢查 HTTP status code
# https://zh.wikipedia.org/zh-tw/HTTP%E7%8A%B6%E6%80%81%E7%A0%81
print(resp.status_code)
```

```python
# 查看下載內容
print(resp.text)
```

```python
# 透過 lxml 解析 HTML
tree = lxml.html.fromstring(resp.text)
```

- XPATH
![XPATH](./screenshot/xpath.png)

```python
xpath = '/html/body/div[3]/div[1]/div/p[1]'
```

```python
data = tree.xpath(xpath)  # 獲取 xpath 位置的資料
```

```python
data[0].text
```
